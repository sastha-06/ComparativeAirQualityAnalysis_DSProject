```python
#from google.colab import files
#uploaded=files.upload()
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("cleaned_city_day.csv")

# 2. Feature Selection
# (Only important columns for AQI comparison)
df = df[['City',
         'PM2.5','PM10',
         'NO2','SO2','CO','O3',
         'AQI_Bucket']]

# 3. Encoding

# Ordinal Encoding for AQI Bucket (Correct CPCB Order)
aqi_order = {
    'Good': 0,
    'Satisfactory': 1,
    'Moderate': 2,
    'Poor': 3,
    'Very Poor': 4,
    'Severe': 5
}

df['AQI_Bucket'] = df['AQI_Bucket'].map(aqi_order)
# 5. Feature Scaling (Z-score)
scaler = StandardScaler()

num_cols = ['PM2.5','PM10','NO2','SO2','CO','O3']

df[num_cols] = scaler.fit_transform(df[num_cols])

# 6. Final Dataset

df.to_csv("final_preprocessed_city_day.csv", index=False)

print("Preprocessing Completed Successfully")
df.head(10)


```
