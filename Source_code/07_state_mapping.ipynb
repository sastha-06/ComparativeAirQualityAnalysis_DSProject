```python
#from google.colab import files
#uploaded = files.upload()
import pandas as pd

# Load datasets
df = pd.read_csv("cleaned_city_day.csv")
city_to_state = {
    'Ahmedabad': 'Gujarat',
    'Aizawl': 'Mizoram',
    'Amaravati': 'Andhra Pradesh',
    'Amritsar': 'Punjab',
    'Bengaluru': 'Karnataka',
    'Bhopal': 'Madhya Pradesh',
    'Brajrajnagar': 'Odisha',
    'Chandigarh': 'Chandigarh',
    'Chennai': 'Tamil Nadu',
    'Coimbatore': 'Tamil Nadu',
    'Delhi': 'Delhi',
    'Ernakulam': 'Kerala',
    'Gurugram': 'Haryana',
    'Guwahati': 'Assam',
    'Hyderabad': 'Telangana',
    'Jaipur': 'Rajasthan',
    'Jorapokhar': 'Jharkhand',
    'Kochi': 'Kerala',
    'Kolkata': 'West Bengal',
    'Lucknow': 'Uttar Pradesh',
    'Mumbai': 'Maharashtra',
    'Patna': 'Bihar',
    'Shillong': 'Meghalaya',
    'Talcher': 'Odisha',
    'Thiruvananthapuram': 'Kerala',
    'Visakhapatnam': 'Andhra Pradesh'
}

df['State'] = df['City'].map(city_to_state)

# Step 2: Create "Others" column
df['Others'] = df[['NO', 'Benzene', 'Toluene', 'Xylene', 'NOx', 'NH3']].sum(axis=1)

# Step 3: Drop unwanted columns
df = df.drop(columns=['City', 'NO', 'Benzene', 'Toluene', 'Xylene', 'NOx', 'NH3'])

df = df[['State', 'Date', 'PM2.5', 'PM10', 'NO2', 'CO', 'SO2', 'O3', 'Others', 'AQI', 'AQI_Bucket']]
# Convert Date to datetime for proper sorting
df['Date'] = pd.to_datetime(df['Date'])

# Sort by State alphabetically and Date ascending
df = df.sort_values(by=['State', 'Date']).reset_index(drop=True)

# Step 4: Final dataset
print(df.head(15))
```
