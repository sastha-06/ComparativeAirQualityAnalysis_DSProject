```python
#from google.colab import files
#uploaded = files.upload()

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_raw = pd.read_csv("city_day.csv")
print("Shape:", df_raw.shape)
df_raw.info()

print("\nStatistical Summary")
print(df_raw.describe())

print("\nMissing Values")
print(df_raw.isnull().sum())

# Numerical Columns
num_raw = df_raw.select_dtypes(include="number")

# Distribution Plots
plt.figure()
plt.hist(df_raw["PM2.5"].dropna())
plt.title("PM2.5 Distribution (Before Cleaning)")
plt.show()

plt.figure()
plt.hist(df_raw["PM10"].dropna())
plt.title("PM10 Distribution (Before Cleaning)")
plt.show()

plt.figure()
plt.hist(df_raw["AQI"].dropna())
plt.title("AQI Distribution (Before Cleaning)")
plt.show()

# Boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data=num_raw)
plt.xticks(rotation=90)
plt.title("Boxplot Before Cleaning")
plt.show()

# Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(num_raw.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Before Cleaning")
plt.show()
```
