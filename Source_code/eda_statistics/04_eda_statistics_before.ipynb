```python
#from google.colab import files
#uploaded = files.upload()

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load cleaned dataset
df_clean = pd.read_csv("cleaned_city_day.csv")

print("Shape After Cleaning:", df_clean.shape)

num_clean = df_clean.select_dtypes(include="number")
pollutants = ['PM2.5','PM10','NO2','SO2','CO','O3','AQI']

for col in pollutants:

    plt.figure(figsize=(6,4))

    sns.histplot(df_clean[col],kde=True)

    plt.title(f"{col} Distribution (After Cleaning)")
    plt.xlabel(col)
    plt.ylabel("Frequency")

    plt.show()
# Boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data=num_clean)
plt.xticks(rotation=90)
plt.title("Boxplot After Cleaning")
plt.show()

# Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(num_clean.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation After Cleaning")
plt.show()
```
